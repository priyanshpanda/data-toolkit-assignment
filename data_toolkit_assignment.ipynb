```python
' Write a Pandas script to find the mean of a specific column in a DataFrameA'
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85, 90, 95]
}

df = pd.DataFrame(data)

mean_score = df['Score'].mean()

print("Mean of Score column:", mean_score)
```

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Sample DataFrame
data = {
    'Math': [85, 90, 78, 92, 88],
    'Science': [80, 85, 75, 95, 90],
    'English': [78, 82, 74, 88, 85]
}

df = pd.DataFrame(data)

# Step 2: Calculate correlation matrix
corr_matrix = df.corr()

# Step 3: Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Step 4: Show the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np

# Step 1: Create two NumPy arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])

# Step 2: Perform element-wise multiplication
result = array1 * array2

# Step 3: Print the result
print("Array 1:", array1)
print("Array 2:", array2)
print("Element-wise multiplication:", result)
```

```python
# Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 3, 4, 2, 1]

# Create line plot with multiple lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', marker='s')
plt.plot(x, y3, label='Line 3', color='red', marker='^')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

# Step 1: Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 67, 92, 74]
}
df = pd.DataFrame(data)

# Step 2: Filter rows where 'Score' > 75
filtered_df = df[df['Score'] > 75]

# Show the filtered DataFrame
print(filtered_df)
```

```python
# Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [55, 60, 65, 70, 70, 75, 75, 80, 85, 85, 90, 95, 100, 100, 100]

# Create histogram with Seaborn
sns.histplot(data, bins=8, kde=True, color='skyblue')

# Add title and labels
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
# Perform matrix multiplication using NumPy
import numpy as np

# Step 1: Create two 2D arrays (matrices)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Step 2: Perform matrix multiplication
result = np.dot(A, B)  # or use A @ B in Python 3.5+

# Step 3: Print the result
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```

```python
# Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [10, 15, 13, 17, 20],
    'z': [5, 6, 7, 8, 9],
    'label': ['A', 'B', 'C', 'D', 'E']  # Optional: for hover info
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='label', title="3D Scatter Plot")

# Show the plot
fig.show()
```
